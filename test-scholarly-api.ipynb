```python
import json

with open("data_peneliti.json", "r") as file:
    data_researcher = json.load(file)
    
type(data_researcher[0])
```

```python
# seach callback for search box
def search_callback(keywords : str):
    result = [item for item in data_researcher if item['name'].lower() == keywords.lower()]
    print(result)
```

```python
search_callback("Heri Santoso")
```

```python
import re

def search_by_regex(data_list, pattern):
    # Menggunakan list comprehension untuk mencocokkan pola regex pada key 'name'
    result = [item for item in data_list if re.search(pattern, item['name'], re.IGNORECASE)]
    print(result)
    
```

```python
search_by_regex(data_researcher, "Dr")
```
